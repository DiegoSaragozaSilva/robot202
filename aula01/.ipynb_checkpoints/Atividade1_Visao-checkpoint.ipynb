## Funções úteis

```python
def plot_hists(img, plt):
    """
        Plota o histograma de cada um dos canais RGB
        img - imagem RGB
        plt - objeto matplotlib
    """
    plt.figure(figsize=(20,10)); 
    img_h = img[:,:,0]
    img_s = img[:,:,1]
    img_v = img[:,:,2]
    histo_plot(img_h, "r","R", plt);
    histo_plot(img_s, "g","G", plt);
    histo_plot(img_v, "b","B", plt);

def make_hist(img_255, c, label, plt):
    """ img_255 - uma imagem com 3 canais de 0 até 255
        c a cor do plot
        label - o label do gráfico
        plt - matplotlib.pyplot
    """
    hist,bins = np.histogram(img_255.flatten(),256,[0,256])
    cdf = hist.cumsum()
    cdf_normalized = cdf * hist.max()/ cdf.max()

    # plt.plot(cdf_normalized, color = c)
    plt.hist(img_255.flatten(),256,[0,256], color = c)
    plt.xlim([0,256])
    plt.legend(label, loc = 'upper left')
    plt.plot()

def histo_plot(img, cor, label, plt):
    """
        img - imagem
        cor - cor
        plt - matplotlib.pyplot object

    """
    plt.figure(figsize=(10,5))
    make_hist(img, cor, label, plt)
    plt.show()
    plt.figure(figsize=(10,5))
    plt.imshow(img, cmap="Greys_r")#, vmin=0, vmax=255)    
    plt.title(label)
```

## Atividade 1

Realize as seguintes atividades:

1. O histograma fornecido aproveita bem as faixas de cores? R: Não, ele fica limitado a uma faixa muito pequena de brilho.
2. Crie uma imagem derivada da imagem fornecida que use melhor a faixa dinâmica, ou seja, tenha o valor dos seus pixels distribuídos entre `0` e `255`. 
3. Plote a imagem pedida

```python
# Guardando o menor e o maior valor de cor da imagem apenas para intuido de análise
vmin = np.amin(rintin_gray)
vmax = np.amax(rintin_gray)

print("Os valores dos pixels tem mínimo de {0} e máximo de {1}.".format(vmin, vmax))

# Criando uma cópia da imagem
rintin_gray_tratado = rintin_gray.copy()

# Aplicando a transformação linear em todos os pixels da imagem
for i in range(len(rintin_gray_tratado)):
    for j in range(len(rintin_gray_tratado[0])):
        # A equação da reta que nos dara o novo valor do pixel é: novo_valor = (51/10) * antigo_valor - (204/5)
        rintin_gray_tratado[i, j] = (51/10) * rintin_gray_tratado[i, j] - (204/5)
        
# Visualizando o resultado e seu histograma
hist,bins = np.histogram(rintin_gray_tratado.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.plot(cdf_normalized, color = 'b')
plt.hist(rintin_gray_tratado.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()

plt.imshow(rintin_gray_tratado, cmap="Greys_r")
```

## Atividade 2

Leia com a OpenCV a figura abaixo:

![](cena_canto_sala.jpg)

Separe os canais  R e B das imagens e analise seus histogramas.

1. A partir dos canais R e B e da análise do histograma é possível identificar a caixa vermelha e o creeper azul?

1. Crie uma imagem numpy em que somente os pixels da caixa vermelha estão brancos e faça seu plot aqui. Você vai precisar usar manipulação de canais RGB. Use o [ExemploFor.ipynb](ExemploFor.ipynb) .


```python
# Carregando a imagem
sala = cv2.imread("cena_canto_sala.jpg")
sala = cv2.cvtColor(sala, cv2.COLOR_BGR2RGB)

# Fazendo a visualização dos canais e dos histogramas
plot_hists(sala, plt)
```

```python
"""
    Analisando os histogramas e as imagens dos canais separadamente, podemos ver que dentro do canal B (azul)
    a caixa possui uma cor muito mais escura que quase todos os elementos presentes na sala, assim, filtrarei
    valores que estão próximos a preto para destacarmos a caixa de branco.
"""
# Criando uma cópia da imagem para fazermos as alterações
sala_caixa = sala.copy()

# Threshold que guardará um valor de margem para que possamos trabalhar com ranges de cores e a cor alvo, no caso, preto.
# Esses dois valores podem ser ajustados para termos resultados melhores, mas, os que estão sendo usado parecem ter resultado satisfatório.
threshold = 10
cor_alvo = 0

# Fazendo o loop pelos pixel e verificando se a cor do pixel está dentro do range definido
for i in range(len(sala_caixa)):
    for j in range(len(sala_caixa[0])):
        if(sala_caixa[i, j, 2] >= cor_alvo - threshold and sala_caixa[i, j, 2] <= cor_alvo + threshold):
            sala_caixa[i, j] = 255
            
# Visualizando o resultado
plt.imshow(sala_caixa)
```

Abra a imagem `hall_box_battery_atividade3.png`  e a converta para tons de cinza de `0` a `255`.

Forneça como saída uma uma cópia da imagem  em que cada pixel  recebe o módulo da subtração entre o pixel que vem depois e o que vem antes dele (na horizontal).

Obs.: O conceito de *antes* e *depois* pode ser na horizontal ou na vertical

Apresente a imagem resultado, deve ficar parecida com a imagem `gabarito_atividade_3.png`.

<img src="gabarito_atividade_3.png" width=400>

```python
# Carregando a imagem em tons de cinza
hall = cv2.imread("hall_box_battery_atividade3.png")
hall = cv2.cvtColor(hall, cv2.COLOR_BGR2GRAY)

# Visualizando a imagem
plt.imshow(hall, cmap="Greys_r", vmin=0, vmax=255)
hall.shape
```

```python
# Criando uma cópia da imagem para ser modificada
hall_mod = hall.copy()

# Aplicando a transformação para cada pixel
for i in range(len(hall)):
    for j in range(len(hall[0])):
        # Previnindo casos de bordas
        if j > 0 and j < len(hall[0]) - 1:
            hall_mod[i, j] = abs(int(hall[i, j + 1]) - int(hall[i, j - 1]))
            
# Visualizando o resultado
plt.imshow(hall_mod, cmap="Greys_r", vmin=0, vmax=255)
```
